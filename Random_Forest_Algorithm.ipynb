***Random Forest Algorithm***
1. Based on supervised
2. Used for both regression and classification problems
3. collection of multiple random decision trees, which is called the forest

```python
# import libraries
from sklearn import datasets
```

```python
# load dataset
iris = datasets.load_iris()
```

```python
iris
```

```python
print(iris.target_names)
```

```python
print(iris.feature_names)
```

```python
import pandas as pd
```

```python
dataset = pd.DataFrame({
    'sepal length': iris.data[:,0],
    'sepal width': iris.data[:,1],
    'petal length': iris.data[:,2],
    'petal width': iris.data[:,3],
    'species': iris.target
})
```

```python
dataset.head()
```

```python
# seeparate dependent and independent variable
X = dataset[['sepal length', 	'sepal width', 	'petal length', 	'petal width']] #features
y = dataset['species'] #labels
```

```python
X
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators  = 100, criterion = "gini")
```

```python
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
X_test
```

```python
clf.score(X_test, y_test)
```

```python
feature_imp = pd.Series(clf.feature_importances_, index = iris.feature_names).sort_values(ascending = False)
```

```python
feature_imp
```
