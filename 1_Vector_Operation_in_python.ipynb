```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1)
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-3,-6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

Addition of two vectors

```python
vector_1 = np.asanyarray([0,0,2,3])

vector_2 = np.asanyarray([0,0,3,-2])
sum = vector_1 + vector_2

print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.quiver(0,0,5,1, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

Substraction of two vector

```python
vector_1 = np.asanyarray([0,0,2,3])

vector_2 = np.asanyarray([0,0,3,-2])
substraction = vector_1 - vector_2

print(substraction)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

Multiplying a vector by scaler

```python
vector_1 = np.asanyarray([0,0,2,3])

vector_2 = 2*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,4,6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```

```python
vector_1 = np.asanyarray([0,0,2,3])

vector_2 = -0.5*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.show()
```
