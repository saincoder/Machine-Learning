***Imbalanced Dataset:***

A dataset with an unequal class distribution

```python
#Import the dependencies
import numpy as np
import pandas as pd
```

```python
#Load the dataset to pandas DataFrame
credit_card_dataset = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_dataset.head()
```

```python
credit_card_dataset.tail()
```

```python
# distruibution of two classes
credit_card_dataset['Class'].value_counts()
```

This is highly imbalanced dataset

0 --> Legit transcation

1 --> Fraudulent transcation

```python
# Separating the Legit and fraudulent teanscation
legit = credit_card_dataset[credit_card_dataset.Class == 0]
fraud = credit_card_dataset[credit_card_dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under Simpling

Building a sample dataset containing similar distribution of legit and raudulent teanscation

```python
legit_sample = legit.sample(n=2)
```

```python
print(legit_sample.shape)
```

```python
new_data_set = pd.concat([legit_sample, fraud],axis = 0)
```
